### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
for num in numeric:
    print(num, ':')
    print('\tAverage:', df[num].mean())
    print('\tStandard deviation:', df[num].std())
    print('\tMinimum:', df[num].min())
    print('\tMaximum:', df[num].max())
    print('\tMode:', df[num].mode()[0])
    print('\tRange:', df[num].max() - df[num].min())
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df.Age, bins=8, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

### 10) A boxplot can also be used to show the distribution of values for each attribute.

```python
numeric_types = df.select_dtypes(include="number")

ax = numeric_types.boxplot()

ax.set_yticks([0, 100, 200, 300, 400, 500, 600, 700, 800, 900])

plt.show()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
for i in range(5):
    df.plot.scatter(numeric[i], numeric[(i+1)%5])
    plt.show()
```
